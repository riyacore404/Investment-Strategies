### Fetch Stock Data

Import the libraries

```python
import pandas as pd
import yfinance as yf
import requests
```

Fetch Stock Information Using yFinance

```python
#fetching csv for top50_us_stocks_info from S&P 500

# Scrape S&P 500 list from Wikipedia
url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
headers = {"User-Agent": "Mozilla/5.0"}
html = requests.get(url, headers=headers).text

# Read all HTML tables on the page
table = pd.read_html(html)

# The first table contains the company data
sp500 = table[0]

# Extract the first 500 stock tickers
top50_tickers = sp500['Symbol'].head(500).tolist()

# Fetch stock info for each ticker using yfinance
data = []
for ticker in top50_tickers:
    stock = yf.Ticker(ticker)
    info = stock.info
    
    # Append essential info about each company to a list
    data.append({
        "Name": info.get("shortName"),
        "Ticker": ticker,
        "Sector": info.get("sector"),
        "MarketCap": info.get("marketCap"),
        "PreviousClose": info.get("previousClose"),
        "52WeekHigh": info.get("fiftyTwoWeekHigh"),
        "52WeekLow": info.get("fiftyTwoWeekLow")
    })

# Step 3: Convert the list into a DataFrame
df = pd.DataFrame(data)

# Save to CSV
df.to_csv("top50_us_stocks_info.csv", index=False)
```
